```python
from glob import glob
import trimesh
import pyvista as pv

shape_paths = glob('../dataset/raw/8867/textured_objs/*')

plotter = pv.Plotter()

for idx, obj in enumerate(shape_paths):
    if obj.endswith('.mtl'):
        continue
    plotter.add_mesh(pv.read(obj), color='grey')


plotter.show()


```
